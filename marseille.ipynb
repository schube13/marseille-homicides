```python
#Arrondissements de Marseille with geoJason
import geopandas as gpd
from shapely.geometry import Point

# 1. Arrondissements laden
arr_df = gpd.read_file("quartiers-marseille.geojson") 
```

```python
# 1. Split Lat/Lon aus Koordinaten_y
df[['Lat', 'Lon']] = df['Koordinaten_y'].str.split(',', expand=True)
df['Lat'] = df['Lat'].astype(float)
df['Lon'] = df['Lon'].astype(float)

# 2. Jahr aus Datum extrahieren
df['Jahr'] = df['Datum'].dt.year

# 3. GeoDataFrame erstellen
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df['Lon'], df['Lat']),
    crs="EPSG:4326"
)

# 4. CRS ggf. angleichen
if arr_df.crs != gdf.crs:
    arr_df = arr_df.to_crs(gdf.crs)

# 5. Spatial Join: Morde mit Quartieren verbinden
gdf_joined = gpd.sjoin(gdf, arr_df, how="left", predicate="within")

# 6. Spaltennamen vereinheitlichen
gdf_joined = gdf_joined.rename(columns={'NOM_CO': 'Arrondissement', 'NOM_QUA': 'Quartier'})

# 7. Finale Auswahl der gewünschten Spalten
gdf_final = gdf_joined[['Datum', 'Text', 'Ort', 'Lat', 'Lon', 'Jahr', 'Arrondissement', 'Quartier']].copy()

# 8. Ergebnis anzeigen
gdf_final.head(10)
```

```python
# 1. Mapping mit korrigierter Schreibweise
arr_to_plz = {
    'Marseille 1er Arrondissemen': '13001',
    'Marseille 2e Arrondissemen':  '13002',
    'Marseille 3e Arrondissemen':  '13003',
    'Marseille 4e Arrondissemen':  '13004',
    'Marseille 5e Arrondissemen':  '13005',
    'Marseille 6e Arrondissemen':  '13006',
    'Marseille 7e Arrondissemen':  '13007',
    'Marseille 8e Arrondissemen':  '13008',
    'Marseille 9e Arrondissemen':  '13009',
    'Marseille 10e Arrondissemen': '13010',
    'Marseille 11e Arrondissemen': '13011',
    'Marseille 12e Arrondissemen': '13012',
    'Marseille 13e Arrondissemen': '13013',
    'Marseille 14e Arrondissemen': '13014',
    'Marseille 15e Arrondissemen': '13015',
    'Marseille 16e Arrondissemen': '13016',
}

# 2. Mapping anwenden (nach Bereinigung)
gdf_final['Arrondissement_clean'] = gdf_final['Arrondissement'].str.strip()
gdf_final['PLZ'] = gdf_final['Arrondissement_clean'].map(arr_to_plz)
```

```python
import numpy as np

# Stärke des Jitters (in Grad)
jitter_strength = 0.0002

# Zähle doppelte Koordinaten (inkl. Wiederholungen)
dups = gdf_joined.duplicated(subset=['Lat', 'Lon'], keep=False)

# Erzeuge neue Spalten
gdf_joined['Lat_jitter'] = df['Lat']
gdf_joined['Lon_jitter'] = df['Lon']

# Nur Duplikate erhalten Jitter
gdf_joined.loc[dups, 'Lat_jitter'] += np.random.uniform(-jitter_strength, jitter_strength, size=dups.sum())
gdf_joined.loc[dups, 'Lon_jitter'] += np.random.uniform(-jitter_strength, jitter_strength, size=dups.sum())
```

```python
from shapely.geometry import Point
import geopandas as gpd

gdf_jitter = gpd.GeoDataFrame(gdf_joined, geometry=gpd.points_from_xy(gdf_joined['Lon_jitter'], gdf_joined['Lat_jitter']), crs='EPSG:4326')
gdf_jitter.to_file("morde_version_jitter.geojson", driver="GeoJSON")
```

```python
# Export als CSV
gdf_jitter.to_csv("Morde_sauber_Marseille.csv", index=False)

```
